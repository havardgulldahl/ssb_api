## Eksempel - grunnkretstabeller i statistikkbanken 

https://www.ssb.no/offentlig-sektor/kommunetall/grunnkretstabeller

```python
# proxy server, base url and grunnkrets table numbers
proxies = {'http': 'http://myProxy:myPort', 'https': 'https://myProxy:myPort'}
base_url = 'https://data.ssb.no/api/v0/no/table/'
tab_nr = [ '04317', '04362', '10184', '10185', '03534', '03536', '10179', '10181', '04363', '10182', '10183']
```

```python
# create urls for grunnkrets tables
url = []
for tab in tab_nr:
    url.append(base_url + tab)
```

```python
# example query for 04317: Folkemengde, etter grunnkrets (1999 - )
q_04317 = {
  "query": [
    {
      "code": "Grunnkretser",
      "selection": {
        "filter": "all",
        "values": [
          "*"
        ]
      }
    },
    {
      "code": "ContentsCode",
      "selection": {
        "filter": "item",
        "values": [
          "Personer1"
        ]
      }
    },
    {
      "code": "Tid",
      "selection": {
        "filter": "all",
        "values": [
          "*"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat"
  }
}
```

```python
# query 'https://data.ssb.no/api/v0/no/table/04317' with q_04317
data = requests.post(url[0], json = q_04317, proxies = proxies)
results = pyjstat.from_json_stat(data.json(object_pairs_hook = OrderedDict))
```
